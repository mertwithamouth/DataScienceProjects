# Pandas Data Visualization 

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df=pd.DataFrame(np.random.rand(1000,5),columns=["A","B","C","D","E"])
```

```python
df.plot.hexbin(x="A",y="C",gridsize=25,cmap="coolwarm")
```
