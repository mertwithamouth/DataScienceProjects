# Chategorical Plots

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset("tips")
tips.head()
```

```python
import numpy as np
sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.std)
```

```python
sns.countplot(x="sex",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",split=True)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips)
sns.swarmplot(x="day",y="total_bill",data=tips, color="black")
```

```python
sns.catplot(x="day",y="total_bill",data=tips, kind="violin")
```

# Matrix Plot

```python
flights=sns.load_dataset("flights")
```

```python
tips.head()
```

```python
flights.head()
```

```python
#we need to turn this into corr table cause index will cause us trouble for heatmap
tc=tips.corr()
```

```python
sns.heatmap(tc,annot=True)
```

```python
#we need to turn this into pivot table cause index will cause us trouble for heatmap
fp=flights.pivot_table(index="month",columns="year",values="passengers")
```

```python
fp.head()
```

```python
sns.heatmap(fp, cmap="magma",linecolor="white",linewidths=0)
```

```python
sns.clustermap(fp,standard_scale=1)
```

# GRIDS

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
iris["species"].value_counts()
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.violinplot)
```

# FACET GRID

```python
g=sns.FacetGrid(data=tips,col='time', row="smoker")
g.map(sns.distplot,'total_bill')
```
